Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.


Ans.
    Simple Linear Regression (SLR)

Definition:
Simple Linear Regression (SLR) is a statistical method used to study the relationship between two variables ‚Äî

one independent variable (X) (also called predictor or explanatory variable), and

one dependent variable (Y) (also called response or outcome variable).

It tries to model the relationship between these two variables by fitting a straight line (called the regression line) to the observed data.
Mathematical Form:
                                         Y=a+bX+e
Where:

Y = Dependent variable (the variable we want to predict)
X = Independent variable (the variable used for prediction)
a = Intercept (value of Y when X = 0)
b = Slope (change in Y for one unit change in X)
e = Error term (difference between observed and predicted Y)

Purpose of Simple Linear Regression:

Prediction:
To predict the value of one variable (Y) based on the known value of another variable (X).
Example: Predicting sales (Y) based on advertising expenditure (X).

Understanding Relationships:
To determine whether and how strongly two variables are related.
Example: Finding if there‚Äôs a positive or negative relationship between study hours and exam scores.

Trend Estimation:
To identify and represent trends or patterns in data using a straight line.

Decision Making:
Helps organizations or researchers make data-driven decisions by quantifying the effect of one variable on another.

Question 2: What are the key assumptions of Simple Linear Regression?

Ans.
   Key Assumptions of Simple Linear Regression (SLR)

For Simple Linear Regression to give accurate and reliable results, several important assumptions must be satisfied. These assumptions ensure that the estimated regression line truly represents the relationship between the variables.

1. Linearity

The relationship between the independent variable (X) and the dependent variable (Y) must be linear.

This means that a change in X produces a proportional change in Y.

Mathematically,
Y=a+bX+e should represent a straight-line relationship.

‚úÖ Example: As study hours increase, exam marks increase roughly in a straight-line pattern.

2. Independence of Errors

The residuals (errors) should be independent of each other.

There should be no correlation between consecutive error terms.

This is especially important in time series data, where errors may be correlated over time.

‚úÖ Test: Durbin-Watson test is often used to check this assumption.

3. Homoscedasticity (Constant Variance of Errors)

The variance of residuals should be constant across all values of X.

This means that the spread of residuals (errors) should remain the same for small and large values of X.

‚ùå Violation Example: If the spread of errors increases with X, it‚Äôs called heteroscedasticity.

4. Normality of Errors

The residuals (errors) should follow a normal distribution.

This assumption is especially important for hypothesis testing and confidence intervals.

‚úÖ Check: A histogram or Q-Q plot of residuals should show a roughly bell-shaped curve.

5. No (or Minimal) Measurement Error in Independent Variable

The independent variable (X) should be measured accurately.

Errors in measuring X can lead to biased or misleading estimates of the slope (b).

6. No Perfect Multicollinearity

This assumption applies mainly to multiple regression, but in SLR, it means there should be only one independent variable, and it must not be a constant or have zero variance.

Question 5: What is the method of least squares in linear regression?

Ans.

The **method of least squares** is a **statistical technique** used in **linear regression** to find the **best-fitting line** through a set of data points.


### üîπ **Concept:**

In simple linear regression, the relationship between ( X ) and ( Y ) is given by:
[
Y = \beta_0 + \beta_1 X + \varepsilon
]

The goal of the least squares method is to find values of ( \beta_0 ) (intercept) and ( \beta_1 ) (slope) that make the **predicted values** of ( Y ) as close as possible to the **actual values**.

### üîπ **How it works:**

For each data point, we calculate the **error (residual)**:
[
\varepsilon_i = Y_i - (\beta_0 + \beta_1 X_i)
]

Then, we **square** these errors and **add them up**:
[
\text{SSE} = \sum (Y_i - \hat{Y_i})^2 = \sum (Y_i - (\beta_0 + \beta_1 X_i))^2
]

The **least squares method** finds the line that **minimizes** this total squared error (SSE).

### üîπ **Why square the errors?**

* To avoid negative and positive errors canceling each other out.
* To give more weight to larger errors.

### üîπ **Result:**

The line obtained by minimizing the sum of squared errors is called the **line of best fit**.
It provides the most accurate prediction of ( Y ) based on ( X ).



Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Ans.

### üîπ **What is Logistic Regression?**

**Logistic Regression** is a **statistical method** used to **predict a categorical (usually binary)** outcome from one or more independent variables.

It estimates the **probability** that a given input belongs to a particular category (e.g., Yes/No, 0/1, Pass/Fail).

---

### üîπ **Mathematical Form:**

Instead of predicting ( Y ) directly, logistic regression predicts the **probability** ( P ) that ( Y = 1 ):

[
P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}
]

Here:

* ( P(Y=1) ) ‚Üí Probability that the event occurs
* ( \beta_0 ) ‚Üí Intercept
* ( \beta_1 ) ‚Üí Coefficient for the independent variable ( X )
* ( e ) ‚Üí Exponential constant (~2.718)

The output is **always between 0 and 1**, representing probability.

### üîπ **Example:**

Predicting whether a student **passes (1)** or **fails (0)** based on **hours studied**.

### üîπ **Difference between Linear and Logistic Regression:**

| **Feature**        | **Linear Regression**                                   | **Logistic Regression**                                |
| ------------------ | ------------------------------------------------------- | ------------------------------------------------------ |
| **Purpose**        | Predicts a **continuous** outcome (e.g., marks, salary) | Predicts a **categorical** outcome (e.g., yes/no, 0/1) |
| **Output Range**   | Can take **any real value**                             | Output lies between **0 and 1** (as probability)       |
| **Equation Type**  | ( Y = \beta_0 + \beta_1 X )                             | ( P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}} )  |
| **Error Function** | Minimizes **sum of squared errors**                     | Uses **maximum likelihood estimation**                 |
| **Graph Type**     | Produces a **straight line**                            | Produces an **S-shaped (sigmoid) curve**               |

Question 7: Name and briefly describe three common evaluation metrics for regression
models.


Ans.
Here are **three common evaluation metrics** used to assess the performance of **regression models**:

---

### üîπ **1. Mean Absolute Error (MAE)**

[
MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y_i}|
]

* It measures the **average absolute difference** between the **actual values** ((Y_i)) and the **predicted values** ((\hat{Y_i})).
* **Lower MAE** indicates better model performance.
* **Easy to interpret** ‚Äî it shows the average error in the same units as the target variable.

---

### üîπ **2. Mean Squared Error (MSE)**

[
MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
]

* It measures the **average of squared differences** between actual and predicted values.
* **Squaring** penalizes larger errors more heavily.
* Useful when **large errors** are particularly undesirable.

---

### üîπ **3. R-squared (Coefficient of Determination)**

[
R^2 = 1 - \frac{\sum (Y_i - \hat{Y_i})^2}{\sum (Y_i - \bar{Y})^2}
]

* Indicates how well the model explains the **variation in the dependent variable**.
* Values range from **0 to 1**:

  * ( R^2 = 1 ): perfect fit
  * ( R^2 = 0 ): model does not explain any variation
* Higher ( R^2 ) means a **better fit**.



Question 8: What is the purpose of the R-squared metric in regression analysis?

Ans.

### üîπ **Purpose of the R-squared Metric:**

The **R-squared (R¬≤)**, also called the **coefficient of determination**, measures how well the **independent variable(s)** explain the **variation** in the **dependent variable** in a regression model.

---

### üîπ **Formula:**

[
R^2 = 1 - \frac{\sum (Y_i - \hat{Y_i})^2}{\sum (Y_i - \bar{Y})^2}
]

Where:

* ( Y_i ) ‚Üí Actual values
* ( \hat{Y_i} ) ‚Üí Predicted values
* ( \bar{Y} ) ‚Üí Mean of actual values

---

### üîπ **Interpretation:**

* ( R^2 ) ranges from **0 to 1**:

  * **R¬≤ = 1:** Perfect model (all variation explained)
  * **R¬≤ = 0:** Model explains nothing about the variation in data
* Example: If ( R^2 = 0.85 ), it means **85% of the variation** in the dependent variable is explained by the independent variable(s).

---

### üîπ **Purpose in regression analysis:**

1. **Measures goodness of fit** ‚Äî how well the model fits the data.
2. **Explains predictability** ‚Äî shows how much of the outcome is predictable from the input variable(s).
3. **Model comparison** ‚Äî helps compare two or more regression models (higher ( R^2 ) usually means a better model).



Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.


```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X = independent variable (hours studied)
# Y = dependent variable (marks obtained)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print the slope (coefficient) and intercept
print("Slope (Œ≤1):", model.coef_[0])
print("Intercept (Œ≤0):", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Ans.

### üîπ **Interpretation of Coefficients in Simple Linear Regression (Theory)**

In a **Simple Linear Regression model**, the relationship between two variables is represented by the equation:

[
Y = \beta_0 + \beta_1 X + \varepsilon
]

Where:

* ( Y ) = Dependent (response) variable
* ( X ) = Independent (predictor) variable
* ( \beta_0 ) = Intercept
* ( \beta_1 ) = Slope (regression coefficient)
* ( \varepsilon ) = Random error term

### üîπ **1. Intercept (Œ≤‚ÇÄ):**

* It represents the **predicted value of Y when X = 0**.
* In other words, it is the **starting point** of the regression line on the Y-axis.
* Example: If the regression equation for marks vs. study hours is
  [
  \text{Marks} = 30 + 6X
  ]
  Then when **hours studied = 0**, the predicted marks are **30**.

### üîπ **2. Slope (Œ≤‚ÇÅ):**

* It represents the **rate of change** in the dependent variable ( Y ) for a **one-unit change** in the independent variable ( X ).
* It shows the **direction and strength** of the relationship:

  * If ( \beta_1 > 0 ): Positive relationship ‚Üí as ( X ) increases, ( Y ) increases.
  * If ( \beta_1 < 0 ): Negative relationship ‚Üí as ( X ) increases, ( Y ) decreases.
* Example: If ( \beta_1 = 6 ), it means for every **1 additional hour studied**, marks increase by **6 points**.
  
### üîπ **3. Overall Meaning:**

The coefficients ( \beta_0 ) and ( \beta_1 ) together describe the **best-fit line** that minimizes the difference between actual and predicted values of ( Y ).
They help us **understand and quantify** how one variable affects another.


```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X = hours studied
# Y = marks obtained
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([35, 40, 50, 55, 60])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Get coefficients
intercept = model.intercept_
slope = model.coef_[0]

# Print results
print("Intercept (Œ≤0):", intercept)
print("Slope (Œ≤1):", slope)
```
